## Convert noise to the frequency domain

Now, we can think of our noise as a function of time: $n(t)$.  You may be familiar with the Fourier transform, which lets us write $n(t)$ as a sum of sines of different sizes, with all the frequencies thatcould be in the data:$$n(t) = \sum_{f_i} \left[ \tilde{n}_{f_i} \sin(2\pi f_i t + \phi_{f_i}) \right]~.$$The sum goes over all frequencies $f_i$ that could be in the data, $\tilde{n}_{f_i}$ are the numbers telling us how much of that particular frequency is in our data (the amplitude), and $\phi_{f_i}$ are the numbers telling us just when that frequency hits its peak.  (For now, we'll ignore $\phi_{f_i}$ to keep things simpler.)  So, for example, if we record a sound that has a frequency of 440 Hz, then $\tilde n_{\text{440 Hz}}$ will be really big compared to all the other $\tilde{n}_{f_i}$.This conversion of $n(t)$ to $\tilde{n}_{f_i}$ and $\phi_{f_i}$ is called the "Fourier transform". You'll definitely see this again someday.  There's also a very quick way of computing it called the "*fast* Fourier transform" (fft), which you might see again some day.  In the next cell, we calculate the fft of the noise you recorded above, and plot the amplitudes on a logarithmic scale.

```python
### Convert to frequency domainNoiseFFT = Noise.dt * np.fft.fft(Noise.data) # Take the Fourier transform and save as 'NoiseFFT'NoiseFrequencies = np.fft.fftfreq(Noise.N, Noise.dt) # Values of the f_i### Only plot some of the data, because python is slow at this jobPlotPoints = Noise.N/2 # first half of the data (positive frequencies)PlotEvery = 500 # every 500th point for speed### Now do the actual plotting:semilogy(NoiseFrequencies[:PlotPoints:PlotEvery], abs(NoiseFFT[:PlotPoints:PlotEvery]));xlabel("$f_i$ (measured in Hertz)");ylabel("$\tilde{n}_{f_i}$ (measured in arbitrary units)");title("Amplitude of the Fourier transform of Noise");
```

```python
### Parameters of our signalSineGaussianCenter = 5.0 # Center of the function in secondsSineGaussianFWHM = 1.9 # full-width at half-max (FWHM) of the Gaussian in secondsSineGaussianFreq = 80 # Frequency of the sine in HertzSineSweep = SineGaussianFreq/4.0 # Amount of frequency sweep from one side of Gaussian to the other### Define the functionsigma = SineGaussianFWHM / (2*sqrt(2*log(2)))def SweptSineGaussianFunction(t) :    return exp(-(t-SineGaussianCenter)**2/(2*sigma**2)) \        * cos(2*pi*(SineGaussianFreq+SineSweep*(t-SineGaussianCenter)/SineGaussianFWHM)*(t-SineGaussianCenter))### Evaluate the function and save the dataSweptSineGaussian = MatchedFiltering.Waveform(Noise)SweptSineGaussian.data = SweptSineGaussianFunction(Noise.Time())### Now plot it as a function of timeplot(SweptSineGaussian.Time(), SweptSineGaussian.data, lw=0.4);xlabel("$t$ (seconds)");ylabel("Amplitude (arbitrary units)");title("Swept sine-Gaussian signal in the time domain");
```

## Matched filtering explainedBecause our signal is so tiny compared to our noise, we can't just look at the data to see if the signal isin there.  Instead, we have to test if the signal $s$ is in the data $d$.  To do this, we measure the **match** between $d$ and $s$, which is defined roughly as$$ M(d, s) \propto \sum_{f_i} \frac{\tilde{d}_{f_i} \tilde{s}_{f_i} \cos \delta\phi_{f_i}}{\tilde{n}_{f_i}^2} $$This formula is really important, so let's try to understand each piece of it.Here, $\delta \phi_{f_i}$ is the difference in the phases $\phi_{f_i}$ of the Fourier transforms of $d$ and $s$.  If $d$ and $s$ are "in phase" for this frequency, $\delta\phi_{f_i}$ will be small, so the cosinewill be roughly 1, and this term will add to the total match.  If $d$ and $s$ are "out of phase", the cosinewill be roughly -1, and this term will make the match smaller.We also have $\tilde{n}_{f_i}$, which is the Fourier-transform amplitude of the noise in the detector, which we've measured previously.  If there's a lot of noise in the detector at a particular frequency, then the data $\tilde{d}_{f_i}$ will be large, but garbage &mdash; so we want to ignore it.  And the louder the noise, the more we want to ignore that frequency.  Dividing each term of the sum by the noise (squared) doesthis automatically for us.  We might think of this like the controls on a stereo.  If the bass is reallyloud and noisy, we might turn it down so we can hear the other parts.After adding up all the terms in the sum, if the match is big, we might say that the signal $s$ is probably in the data $d$.  And if the match is small, we probably didn't detect the signal.  In fact, if the data andthe signal are a perfect match, then$$\text{Match}(d,s) = \frac{M(d,s)} {\sqrt{M(d,d) M(s,s)}} $$will be 1.  So this is the quantity we're most interested in.Once you've read this far, take the time to ask any questions and discuss this concept.
