## Another way of looking at the noise: the frequency domain

Now, we can think of our noise as a function of time: $n(t)$.  You may be familiar with the Fourier transform, which lets us write $n(t)$ as a sum of sines of different sizes, with all the frequencies thatcould be in the data:$$n(t) = \sum_{f_i} \left[ \tilde{n}_{f_i} \sin(2\pi f_i t + \phi_{f_i}) \right]~.$$The sum goes over all frequencies $f_i$ that could be in the data, $\tilde{n}_{f_i}$ are the numbers telling us how much of that particular frequency is in our data (the amplitude), and $\phi_{f_i}$ are the numbers telling us just when that frequency hits its peak.  (For now, we'll ignore $\phi_{f_i}$ to keep things simpler.)  So, for example, if we record a sound that has a frequency of 440 Hz, then $\tilde n_{\text{440 Hz}}$ will be really big compared to all the other $\tilde{n}_{f_i}$.This conversion of $n(t)$ to $\tilde{n}_{f_i}$ and $\phi_{f_i}$ is called the "Fourier transform". You'll definitely see this again someday.  There's also a very quick way of computing it called the "*fast* Fourier transform" (fft), which you might see again some day.  In the next cell, we calculate the fft of the noise you recorded above, and plot the amplitudes on a logarithmic scale.

```python
### Convert to frequency domainNoiseFFT = Noise.dt * np.fft.fft(Noise.data) # Take the Fourier transform and save as 'NoiseFFT'NoiseFrequencies = np.fft.fftfreq(Noise.N, Noise.dt) # Values of the f_i### Now plot it:figure()PlotPoints = Noise.N/2 # show first half of the data (positive frequencies)#semilogy(NoiseFrequencies[:PlotPoints:PlotEvery], abs(NoiseFFT[:PlotPoints:PlotEvery]));#xlim(0,25000);xlim(0,5000);ylim(ymin=1e-4);xlabel("$f_i$ (measured in Hertz)");ylabel("$\\tilde{n}_{f_i}$ (measured in arbitrary units)");title("Amplitude of the Fourier transform of Noise");
```

Your line probably jumps around a bit, with lots of spikes up and down, but generallydecreasing with higher frequencies.  This means that there's a lot of noise at low frequencies (bass), and less at high frequencies (treble).

<br /><br /><br /># Create a simple signal, inject it, and extract itHere's where we'll make up a signal that's kind of vaguely like a real gravitational-wave signal, but is really easy to calculate.  Then, we'll add it to our noise data &mdash; but so quietly that we can't hear it.Finally, we'll detect it again with matched filtering.## Make up a "swept sine-Gaussian" signalThis will be a roughly sinusoidal signal with gradually increasing (sweeping) frequency, but its amplitude is a Gaussian function.

```python
### Set the parameters of our signalSineGaussianCenter = 5.0 # Center of the function in secondsSineGaussianFWHM = 2.4 # full-width at half-max (FWHM) of the Gaussian in secondsSineGaussianFreq = 1200 # Frequency of the sine in HertzSineSweep = SineGaussianFreq/20.0 # Amount of frequency sweep across the Gaussian FWHM### Define the functionsigma = SineGaussianFWHM / (2*sqrt(2*log(2)))def SweptSineGaussianFunction(t) :    return exp(-(t-SineGaussianCenter)**2/(2*sigma**2)) \        * cos(2*pi*(SineGaussianFreq+SineSweep*(t-SineGaussianCenter)/SineGaussianFWHM)              *(t-SineGaussianCenter))### Evaluate the function and save the dataSweptSineGaussian = MatchedFiltering.Waveform(Noise)SweptSineGaussian.data = SweptSineGaussianFunction(Noise.Time())### Now plot it as a function of timefigure()plot(SweptSineGaussian.Time(), SweptSineGaussian.data, lw=0.4);gcf().set_size_inches(9,5);xlabel("$t$ (seconds)");ylabel("Amplitude (arbitrary units)");title("Swept sine-Gaussian signal in the time domain");
```

## Matched filtering explainedBecause our signal is so tiny compared to our noise, we can't just look at the data to see if the signal isin there.  Instead, we have to test if the signal $s$ is in the data $d$.  To do this, we measure the **"match"** between $d$ and $s$, which filters the data using our knowledge of the expected signal and thetypical noise in the detector, and then measures the total power in the filtered result.  If there's a lotof power, then the signal probably really was in our data.We start out with the amplitude of the Fourier transform of the data, $\tilde{d}_{f_i}$, and let's just thinkof one particular frequency for now.  If our expected signal $\tilde{s}_{f_i}$ is small, we want to ignore this part of the data; if $\tilde{s}_{f_i}$ is large, we want to accentuate this part of the data.  So thefirst step in our filter is to take the product$$\tilde{d}_{f_i} \tilde{s}_{f_i}~.$$If the signal is really in our data, then this picks out the meaningful part of that data.On the other hand, if the detector's noise at this frequency is large, we expect the data to be large, butuseless garbage.  So we want to ignore the data whenever there's a lot of noise.  To do that, we divide by the noise at this frequency $\tilde{n}_{f_i}$.  But we don't want to just level the playing field for frequencies with lots of noise; we want to really eliminate that bad data &mdash; so we divide by *two* powers of the noise:$$ \frac{\tilde{d}_{f_i} \tilde{s}_{f_i}} {\tilde{n}_{f_i}^2}~. $$We might think of this like the controls on a stereo.  If the bass is really loud and noisy, we might turn it down so we can hear the other parts.This isn't quite the whole story, though, because the Fourier transform really involves two pieces of information: the amplitude and the phase.  The phase $\phi_{f_i}$ tells us exactly where the sine with frequency $f_i$ hits its peak.  And if $d$ hits its peak when $s$ hits a trough, we don't really have the same signal.  So we bring in one more factor to our filter:$$ \frac{\tilde{d}_{f_i} \tilde{s}_{f_i}} {\tilde{n}_{f_i}^2} \cos \delta\phi_{f_i}~, $$where $\delta \phi_{f_i}$ is the difference in the phases $\phi_{f_i}$ of the Fourier transforms of $d$ and $s$ at this frequency.  Now, for each $f_i$, the quantity above is large whenever the noise is small and the expected signal agrees with the observed data &mdash; as long as our expected signal is really in the data.We want just one number to tell us whether the entire signal (for all frequencies) really agrees with the data, so we just sum the expression above for all frequencies we measured:$$ M(d, s) \propto \sum_{f_i} \frac{\tilde{d}_{f_i} \tilde{s}_{f_i}}{\tilde{n}_{f_i}^2} \cos \delta\phi_{f_i}~. $$This is called the "match" (up to factors of two or so), and it's the key quantity we measure when trying to detect gravitational waves.Once you've read this far, take the time to ask any questions and discuss this concept to make sure youunderstand what's going on.

## Matched filtering in actionNow that you understand matched filtering, let's try it out with our fake data.  We'll show the match quantity as a function of time offset, first between the signal and the plain old noise (so we expectmore noise as a result), then between the signal and our fake detector data.

```python
## Calculate the match with the detector data and just plain noise (no signal)MatchWithPlainNoise = MatchedFiltering.Match(Noise, Signal, Noise)MatchWithDetectorData = MatchedFiltering.Match(DetectorData, Signal, Noise)### Compare peak matchesprint("Largest match with plain noise   = {}".format(max(MatchWithPlainNoise[50:-50])))print("Largest match with detector data = {}".format(max(MatchWithDetectorData[50:-50])))print("Output signal-to-noise ratio = {}".format(max(MatchWithDetectorData[50:-50])                          /max(MatchWithPlainNoise[50:-50])))print("Input signal-to-noise ratio = {}".format(InputSignalToNoiseRatio))### Plot the resultsfigure()semilogy(Noise.Time()[::PlotEvery], MatchWithDetectorData[::PlotEvery],          label="$\\mathrm{Match}(d,s)$");semilogy(Noise.Time()[::PlotEvery], MatchWithPlainNoise[::PlotEvery],          label="$\\mathrm{Match}(n,s)$");ylim(1e-4, 1);xlabel("Time offset (seconds)");ylabel("Match");legend(loc="upper right");
```

This plot shows that if we search for the signal in pure noise, we get noise &mdash; a bunch of spikes, butnone really bigger than any other.
